# import modules and make display smaller

```python
import pandas as pd

# Make display smaller
pd.options.display.max_rows = 10
```

# read the data from users, ratings and movie names files

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('users.dat', sep='::',
                      engine='python', header=None, names=unames)
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ratings.dat', sep='::',
                        engine='python', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::',
                       engine='python', header=None, names=mnames)
```

# merge all the three files in the following way

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

# display the ratings by title

```python
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title
```

# display first 10 rows of the ratings by title

```python
ratings_by_title[:10]
```

# find and display the active titles of movies

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
```

# find the mean ratings for those active movie titles

```python
mean_ratings = data.pivot_table('rating', index='title',
                                columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```python
# Select rows on the index
mean_ratings = mean_ratings.loc[active_titles]
mean_ratingmean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

# find the top movie ratings given by ladies

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

# find top movie ratings by gentlemen

```python
top_male_ratings = mean_ratings.sort_values(by='M', ascending=False)
top_male_ratings[:10]
```

# find the difference

```python
mean_ratings['diff'] = mean_ratings['F'] - mean_ratings['M']
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:20]
```

```python
len(sorted_by_diff[sorted_by_diff['diff'] > 1])
```

```python
# Reverse order of rows, take first 10 rows
sorted_by_diff[::-1][:10]
```

```python
# Standard deviation of rating grouped by title
rating_std_by_title = data.groupby('title')['rating'].std()
print(rating_std_by_title)
# Filter down to active_titles
rating_std_by_title = rating_std_by_title.loc[active_titles]
print(rating_std_by_title)
# Order Series by value in descending order
rating_std_by_title.sort_values(ascending=False)[:10]
```
